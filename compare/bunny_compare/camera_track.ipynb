```python
import os
import json
import numpy as np
```

```python
with open('exp2_transforms_test.json') as json_file:
    json_data = json.load(json_file)
```

```python
x1 = np.array([0,4,0.5,1])
```

```python
transform_list = []
```

```python
for i in range(len(json_data['frames'])):
    transform_list.append(json_data['frames'][i]['transform_matrix'])
```

```python
transform_list = np.array(transform_list)
```

```python
c1 = []
```

```python
for i in range(len(json_data['frames'])):
    k = np.dot(transform_list[i], x1)
    c1.append(k[:3])
```

```python
s = 0
for i in c1[4]:
    s += i**2
```

```python
np.sqrt(s)
```
