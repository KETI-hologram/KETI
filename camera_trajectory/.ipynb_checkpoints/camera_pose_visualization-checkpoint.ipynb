```python
import utility
import numpy as np
import os
import natsort
#import open3d as o3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
%matplotlib inline
```

```python
def rotationMatrixToEulerAngles(R) :
    sy = math.sqrt(R[0,0] * R[0,0] +  R[1,0] * R[1,0])
    singular = sy < 1e-10
    if  not singular :
        x = math.atan2(R[2,1] , R[2,2])
        y = math.atan2(-R[2,0], sy)
        z = math.atan2(R[1,0], R[0,0])
    else :
        x = math.atan2(-R[1,2], R[1,1])
        y = math.atan2(-R[2,0], sy)
        z = 0
    return np.array([x, y, z])
```

```python
import json

with open("ct/benz/exp1_transforms_test.json", "r") as json_file:
    json_data = json.load(json_file)
frames = json_data['frames']
```

```python
len(frames)
```

```python
position = []
orientation = []
for i in range(len(frames)):
    matrix = np.array(frames[i]['transform_matrix'])[:3]
    r = rotationMatrixToEulerAngles(matrix[:,:3])
    t = matrix[:,3:]
    #d = np.dot(r.T, np.array([0,0,1]))
    position.append(t.T)
    orientation.append(r.T)
positions = np.vstack(position)
orientations = np.vstack(orientation)
```

```python
with open("ct/benz/exp1_transforms_train.json", "r") as json_file:
    train_json_data = json.load(json_file)
train_frames = train_json_data['frames']
```

```python
len(train_frames)
```

```python
train_position = []
train_orientation = []
for i in range(len(train_frames)):
    matrix = np.array(train_frames[i]['transform_matrix'])[:3]
    r = rotationMatrixToEulerAngles(matrix[:,:3])
    t = matrix[:,3:]
    #d = np.dot(r.T, np.array([0,0,1]))
    train_position.append(t.T)
    train_orientation.append(r.T)
train_positions = np.vstack(train_position)
train_orientations = np.vstack(train_orientation)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.set_zlim(-4,4)
ax.plot(positions[:,0],positions[:,1],positions[:,2], color = 'dodgerblue', label='Test')
ax.scatter(train_positions[:,0],train_positions[:,1],train_positions[:,2], color='limegreen', label='Train')
ax.scatter(0,0,0, color = 'red', s=100, label='Object')
ax.text(0, 0, 0, "(0,0,0)", color='black', fontsize = 'xx-large')
ax.quiver(positions[:,0], positions[:,1], positions[:,2], -positions[:,0] , -positions[:,1], -positions[:,2], length=0.5, normalize=True, color='dodgerblue')
plt.title('Camera trajectory Car_EXP1', fontsize=25)
plt.legend()
plt.show()
```
